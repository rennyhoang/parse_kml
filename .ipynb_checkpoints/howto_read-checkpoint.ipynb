# Example reading the time-series and the cumulative displacement files

```python
ts_file = './files/sentinel_dvert_timeseries_median_filtered.h5'
disp_file = './files/ifgram_recon_displacement.h5'
mask_file = './files/mask_low_elevation_high_tempCoh.h5'
```

```python
# define a function that does not rely on mintpy
import h5py
import numpy as np
import pandas as pd
def readtimeseries(filepath):
    # read the data
    f = h5py.File(filepath,'r')
    dset = f['timeseries']
    dset = np.array(dset)
    dates = [i.decode('utf8') for i in f['date'][:]]
    dates = pd.to_datetime(dates)
    f.close()
    return dset, dates

def readdisplacement(filepath):
    # read the data
    f = h5py.File(filepath,'r')
    dset = f['velocity']
    dset = np.array(dset)
    f.close()
    return dset
    
def readmask(filepath):
    # read the data
    f = h5py.File(filepath,'r')
    dset = f['mask']
    dset = np.array(dset)
    f.close()
    return dset
```

```python
def latlon2yx(lat,lon):
    # 33.50 34.4 -118.75 -117.25 
    lat0 = 34.4; lon0= -118.75 
    step = 0.00277778

    y = -int((lat-lat0)/step)
    x = int((lon-lon0)/step)

    return y,x
```

```python
# read time-series 
dset_ts, dates = readtimeseries(ts_file)
# Plot timeseries at a pixel in the center of the SGV
lat = 34.096
lon = -118.030
y,x = latlon2yx(lat,lon)
print('Pixel location: ',y,x)
ts_yx = dset_ts[:,y,x]
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(8,4))
ax.scatter(dates,ts_yx,c='k',label='Sentinel-1')
ax.set_xlabel('Date')
ax.set_ylabel('Displacement [m]')
```

```python
# read cumulative displacement between 2022_12_27 and 2023_06_01
dset_disp = readdisplacement(disp_file)
# read mask
dset_mask = readmask(mask_file)
# plot
fig, ax = plt.subplots(figsize=(8,4))
ax.imshow(dset_disp*dset_mask, cmap='jet',vmin=-0.05,vmax=0.05)

```

```python
# derive the cumulative displacement directly from the time-series
date1 = '2022-12-27'
date2 = '2023-06-01'
idx1 = dates.get_loc(date1)
idx2 = dates.get_loc(date2)
disp_idx1_idx2 = dset_ts[idx2,]-dset_ts[idx1,]
plt.figure()
plt.imshow(disp_idx1_idx2*dset_mask, cmap='jet', vmin=-0.05, vmax=0.05)
```
